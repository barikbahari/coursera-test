```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Upload file dari gdrive, tapi gagal hehe
#from google.colab import drive
#drive.mount('/content/drive')
#filepath = '/content/drive/MyDrive/Colab Notebooks/Filling A1.csv'

# Upload manual
from google.colab import files
uploaded = files.upload()

filling_a1 = pd.read_csv('Filling A1.csv')
filling_a1['decimalBagLength'] = filling_a1['decimalBagLength'] / 10
filling_a1.head()

df = pd.DataFrame(filling_a1)

# Statistik Deskriptif
print("Statistik Deskriptif:\n", df.describe())

# Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, y="intAugerSpeed", color="skyblue")
plt.title("Distribusi Auger Speed", fontsize=14)
plt.ylabel("Auger Speed (bpm?)", fontsize=12)
plt.show()

# Scatter plot Product Weight vs Auger Speed
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="intSamplingProductWeight", y="intAugerSpeed", hue="intIdParameterE1", palette="viridis", s=100)
plt.title("Hubungan Product Weight dan Auger Speed", fontsize=14)
plt.xlabel("Product Weight (mg)", fontsize=12)
plt.ylabel("Auger Speed (bpm)", fontsize=12)
plt.legend(title="ID", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Histogram Auger Speed
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="intAugerSpeed", bins=5, kde=True, color="green")
plt.title("Histogram Auger Speed", fontsize=14)
plt.xlabel("Auger Speed (bpm)", fontsize=12)
plt.ylabel("Frekuensi", fontsize=12)
plt.show()

# Insight tambahan
print("Auger Speed maksimum terjadi pada:", df.iloc[df['intAugerSpeed'].idxmax()])
print("Auger Speed minimum terjadi pada:", df.iloc[df['intAugerSpeed'].idxmin()])
```
